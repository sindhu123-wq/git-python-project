```python
#Step 1
#import data from csv file
import pandas as pd
data=pd.read_csv('50_Startups_MLR.csv')
```

```python
data
```

```python
data.columns
```

```python
#Step 2
#Separate both input and output variables
#Independent Variables

X=data.iloc[:,0:-1].values
```

```python
X
```

```python
#Dependent Variables

Y=data.iloc[:,-1].values
```

```python
Y
```

```python
#Step3-Missing values

#Missing values in X

from sklearn.impute import SimpleImputer
import numpy as np

imputer=SimpleImputer(missing_values=0.0,strategy='mean')
imputer.fit(X[:,0:3])  #identification
X[:,0:3]=imputer.transform(X[:,0:3])
```

```python
X
```

```python
#Step4-Coverting string datatype to int

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
X=ct.fit_transform(X)
```

```python
X
```

```python
#Step 5:Feature scaling
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X[:,0:3]=sc.fit_transform(X[:,0:3])
```

```python
X
```

```python
#Step6 : Splitting the data into train and test

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```
